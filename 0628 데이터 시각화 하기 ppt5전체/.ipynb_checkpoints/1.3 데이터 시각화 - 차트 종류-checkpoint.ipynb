# 차트 종류 변경

```python
#필요 라이브러리 정의
import numpy as np
import matplotlib.pyplot as plt

#데이터 정의 150개의 2열짜리 정규분포 생성
y=np.random.standard_normal((150,2))

plt.figure(figsize=(7,5))
plt.xlabel('1st')
plt.ylabel('2nd')
plt.title('Scatter plt')
plt.scatter(y[:,0],y[:,1],marker='o', edgecolors='r')
#plt.scatter?
```

# 차원추가 (Scatter 차트) - 3차원까지 보여줄 수 있다.

```python
#필요 라이브러리 정의
import numpy as np
import matplotlib.pyplot as plt

#데이터 정의 정규분포 50개행 컬럼2개
y=np.random.standard_normal((50,2))
# 0~10까지 랜덤숫자 50개
c= np.random.randint(0,10,len(y))

plt.figure(figsize=(7,5))
plt.scatter(y[:,0],y[:,1],marker='o',c=c)
plt.colorbar()

plt.xlabel('1st')
plt.ylabel('2nd')
plt.title('Scatter plt')
#plt.scatter?
```

```python
#객체 차트
# 빈캔버스에 추가함
fig = plt.figure()
# 위치 및 너비/높이

# 1~5까지 11개의 점 삽입
x = np.linspace(1,5,11)
y = x**2

ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

ax1.plot(x , y)
ax2.plot(y , x)
ax1.set_title('plot1')
ax2.set_title('plot2')
fig
```

# 객체 차트 (크기/해상도 조정)

```python
#길이크기조정, 해상도 조정

fig = plt.figure(figsize = (8,2), dpi = 60)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
fig
```
