# 2. 조건 판단하기

# if [조건문]: # 조건이 일치하는 경우 실행
#    [조건문 참인경우 실행]
# else:
#    [조건문 불일치 시 실행]


```python
# 변수에 특정 값 할당
stockname = 'SEA'
```

```python
# 조건 식 수행 일치 시 실행
if stockname == "SEC":
    print("Union")
elif stockname == "SEA":
    print("True")
# 조건이 일치하지 않는 경우 else 조건 실행
else:
    print("False")
```

# 3. 반복하기 (while, for)

# while (조건) loop : 조건 실패 시 조건 문 탈출

# for (조건) : 조건 실패 시 조건 문 탈출

```python
#반복문 예제 #1
i = 1
while i <= 10:
    print(i)
    i=i+1;
print("1번문제 시작!")
```

```python
#반복문 예제 #2
for i in range(1, 11):
    print(i)
print("2번문제 시작!")
```

# 2중 for문

```python
for n in [1,2]:
    print("{0}단 시작".format(n))
    for i in [1,2,3,4,5,6,7,8,9]:
        print("{0} * {1} =  {2} ".format(n, i, n*i))
```

```python
for n in [1,2]:
    print("{0}단 시작".format(n))
    for i in [1,2,3,4,5,6,7,8,9]:
        if i > 5:
            break
        print("{0} * {1} =  {2} ".format(n, i, n*i))
```

```python
for n in [1,2]:
    if n%2 == 0:
            continue
    print("{0}단 시작".format(n))
    for i in [1,2,3,4,5,6,7,8,9]:
        print("{0} * {1} =  {2} ".format(n, i, n*i))
```
