```python
import requests, bs4, pandas as pd, numpy as np
 
#웹페이지 html 소스 가져오기
resp = requests.get('http://finance.daum.net/quote/marketvalue.daum')
resp.encoding='utf-8-sig'
html = resp.text
bs = bs4.BeautifulSoup(html, 'html.parser')
# html 부분구조 가져오기
columns=bs.select('ul#myListTop1 > li > a', limit=7)
#columns2=bs.select('ul#myListTop1 > li > span.cost.num', limit=7)
columns2=bs.select('ul#myListTop1 > li ', limit=7)
```

```python

columns
columns2
```

```python
a = ["stock_name"]
columnlist=[]
for column in columns:
    columnlist.append(column.text)
df=pd.DataFrame(data = columnlist, dtype = str)
df
```

```python
columnlist2 =[]

for column in columns2:
    columnlist22 = float(column.find("span",{"class":"cost num"}).text.replace(",", ""))
    columnlist2.append(columnlist22)

df2=pd.DataFrame(data = columnlist2, dtype = str)
df2
```

```python
dff = pd.concat([df, df2], axis=1)
dff
```

```python
dff.columns = ['stock_name','stock_price']
dff
```
